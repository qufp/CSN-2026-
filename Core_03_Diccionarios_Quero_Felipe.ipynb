###1. Crea una lista de diccionarios llamada ventas, donde cada diccionario represente una venta. Cada venta debe incluir las siguientes claves:

* «fecha»: una cadena de texto que represente la fecha de la venta (por ejemplo, «2024-01-01»).
* «producto»: una cadena de texto que represente el nombre del producto vendido.
* «cantidad»: un número entero que represente la cantidad de productos vendidos.
* «precio»: un número flotante que represente el precio unitario del producto.



```python
ventas = [
    {"fecha": "2026-06-01", "producto": "Jeans ", "cantidad": 2, "precio": 60000},
    {"fecha": "2026-06-01", "producto": "Poleron", "cantidad": 2, "precio": 45000},
    {"fecha": "2026-06-02", "producto": "Shorts", "cantidad": 2, "precio": 50000},
    {"fecha": "2026-06-02", "producto": "Zapatillas", "cantidad": 1, "precio": 120000},
    {"fecha": "2026-06-03", "producto": "Gorro", "cantidad": 3, "precio": 90000},
    {"fecha": "2026-06-04", "producto": "Polera", "cantidad": 3, "precio": 50000},
    {"fecha": "2026-06-05", "producto": "Calcetines", "cantidad": 4, "precio": 20000}
]

ingresos_totales = 0

for venta in ventas:
    ingreso_venta = venta["cantidad"] * venta["precio"]
    ingresos_totales = ingresos_totales + ingreso_venta

ventas_por_producto = {}

for venta in ventas:
    producto = venta["producto"]
    cantidad = venta["cantidad"]

    if producto in ventas_por_producto:
        ventas_por_producto[producto] = ventas_por_producto[producto] + cantidad
    else:
        ventas_por_producto[producto] = cantidad

producto_mas_vendido = ""
max_cantidad = 0

for producto, cantidad in ventas_por_producto.items():
    if cantidad > max_cantidad:
        max_cantidad = cantidad
        producto_mas_vendido = producto

precios_por_producto = {}

for venta in ventas:
    producto = venta["producto"]
    ingreso_total_producto = venta["cantidad"] * venta["precio"]
    cantidad = venta["cantidad"]

    if producto in precios_por_producto:
        suma_actual, cant_actual = precios_por_producto[producto]
        precios_por_producto[producto] = (suma_actual + ingreso_total_producto, cant_actual + cantidad)
    else:
        precios_por_producto[producto] = (ingreso_total_producto, cantidad)


ingresos_por_dia = {}

for venta in ventas:
    fecha = venta["fecha"]
    ingreso = venta["cantidad"] * venta["precio"]

    if fecha in ingresos_por_dia:
        ingresos_por_dia[fecha] = ingresos_por_dia[fecha] + ingreso
    else:
        ingresos_por_dia[fecha] = ingreso

resumen_ventas = {}

for producto, datos_tupla in precios_por_producto.items():

    ingreso_total = datos_tupla[0]
    cantidad_total = datos_tupla[1]
    precio_promedio = ingreso_total / cantidad_total

    resumen_ventas[producto] = {
        "cantidad_total": cantidad_total,
        "ingresos_totales": ingreso_total,
        "precio_promedio": precio_promedio
    }

print("=== REPORTE DE ANÁLISIS DE VENTAS ===\n")

print("1. LISTA DE VENTAS ORIGINAL:")
for venta in ventas:
    print(f" - {venta}")

print(f"\n2. INGRESOS TOTALES GENERADOS: ${ingresos_totales}")

print(f"\n3. PRODUCTO MÁS VENDIDO:")
print(f" - Producto: {producto_mas_vendido}")
print(f" - Cantidad: {max_cantidad} unidades")

print("\n4. PRECIO PROMEDIO DE VENTA POR PRODUCTO:")
for producto, datos in resumen_ventas.items():
    print(f" - {producto}: ${datos['precio_promedio']}")

print("\n5. INGRESOS TOTALES POR DÍA:")
for fecha, ingreso in ingresos_por_dia.items():
    print(f" - {fecha}: ${ingreso}")

print("\n6. RESUMEN DE VENTAS POR PRODUCTO:")
for producto, detalles in resumen_ventas.items():
    print(f" - {producto}:")
    print(f"     Cantidad total vendida: {detalles['cantidad_total']}")
    print(f"     Ingresos totales: ${detalles['ingresos_totales']}")
    print(f"     Precio promedio: ${detalles['precio_promedio']}")
```
